```python
x1 = np.array([1.93, -1.05, 0.7, -0.26, 1.44, 1.40, -0.64, 2.14, -2.07, -1.59, -3.42, -1.22, 2.06, 0.53, 1.15, -1.15, -0.23, -1.11, -0.59, 2.32, -3.58, 1.24, 1.82, 2.65, -3.67])
x2 = np.array([-0.85, -2.48, 2.33, -2.84, 0.41, -0.10, 3.38, -2.94, 1.68, 2.42, 1.37, -1.46, 1.64, -0.83, 2.92, 1.46, -2.82, 2.67, -2.43, 0.23, -1.17, -1.38, -3.92, 0.51, -0.12])
x3 = np.array([0.45, 0.37, 1.08, -1.68, 0.71, 0.61, 0.75, 0.49, -0.35, 0.44, -1.97, 1.13, 1.31, 0.45, 0.28, -0.01, -1.79, 0.58, -4.69, 2.75,-3.29, -1.98, 3.54, -0.74, 1.55])
n = len(x1)
Matrix = sp.Matrix([x1,x2,x3])
Matrix 
```

Составим матрицу ковариации

```python
x1x1 = sum(x1*x1)/n
x1x2 = sum(x1*x2)/n
x1x3 = sum(x1*x3)/n
x2x1 = sum(x2*x1)/n
x2x2 = sum(x2*x2)/n
x2x3 = sum(x2*x3)/n
x3x1 = sum(x3*x1)/n
x3x2 = sum(x3*x2)/n
x3x3 = sum(x3*x3)/n
# перемножили векторы и отнормировали 
x3x1
```

Далее из полученных значений составляем матрицу ковариации

```python
matrix = np.array ([[x1x1, x1x2, x1x3],[x2x1, x2x2, x2x3], [x3x1, x3x2, x3x3]])
matrix 
```

Найдем собственные числа

```python
evalue, evect = np.linalg.eig(matrix)

print ("Собственные числа равны:", evalue, sep='\n')
print ("Собственные вектора равны:", evect, sep='\n')
```

```python
m = evect
m
```

Далее выполняем сортировку векторов по собственным числам от большего к меньшему и переставляем собственные вектора в порядке убывания собственных значений и составляем матрицу перехода

```python
m1 = evect [:,0] #срез по столбцам
m2 = evect [:,1]
m3 = evect [:,2]

T = np.column_stack([m3,m2,m1])
T
```

Находим обратную матрицу перехода

```python
T1 = np.linalg.inv (T)
T1
```

Вычисляем координаты всех выборочных точек в пространстве Y1,Y2,Y3
