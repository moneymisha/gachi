Создание графика функции

```python
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d.axes3d import Axes3D
import numpy as np
import sympy
from sympy import diff
import matplotlib.pyplot as plt

n = 1
x = np.linspace(-10, 10, 50) #значения х (наим, наиб, число точек)
y = n*x**3 + n**2*x**2 +n**3*x+n**4 
y1 = 3*n*x**2 + 2*n*2*x + 3*n**2*x+4*n**3 #производная
y2 = 6*n*x + 4*n*x + 6*n*x + 12*n**2 #2 производная

fig, ax = plt.subplots () #фигура и оси
plt.plot(x, y, color = "blue", label = "y(x)") 
plt.plot(x, y1, color = "red", label = "y'(x)") #cтроим график производной (аргумент, значение, цвет, подпись)
plt.plot(x, y2, color = "yellow", label = "y''(x)")
ax.set_xlabel ("x") #ось x
ax.set_ylabel ("y") #ось y
ax.grid(True) #добавление сетки
plt.legend ()
```

Создание фигуры

```python
fig = plt.figure (figsize=(10,10)) #создание осей
ax_3d = fig.add_subplot(projection='3d')

a = np.linspace (0, 2*np.pi, 100) #создание массива от 0 до 2 пи с шагом 100
b = np.linspace (0,np.pi,100)

x = 10 * np.outer(np.cos(a), np.sin(b)) #произведение 2-х векторов
y = 10 * np.outer(np.sin(a), np.sin(b))
z = 10 * np.outer(np.ones(np.size(a)), np.cos(b))

ax_3d.plot_surface(x, y, z, rstride=4, cstride=4, color='green')

plt.show()

```

```python
fig = plt.figure (figsize=(10,10)) #создание осей
ax_3d = fig.add_subplot(projection='3d')

a = np.linspace (0, 100, 50) 
b = np.linspace (0, 100, 50)

x = 5*np.outer(a,a)
y = 5* np.outer(b,b)
z = np.outer (x,y)

ax_3d.plot_surface(x, y, z, rstride=10, cstride=14, color='green')

plt.show()
```
